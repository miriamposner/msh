```python
import sys
from pymarc import MARCReader
with open('msh20160604-marc8.mrc', 'rb') as fh:
    reader = MARCReader(fh)
    print('maori word; use instead of; related term; broader term')
    for record in reader:
        maori, instead, related, broader = [], [], [], []
        for field in record.get_fields():
            if field.tag == '150':
                maori.append(field.value())
            elif field.tag == '450':
                instead.append(field.value())
            elif field.tag == '550':
                if 'g' in field.subfields:
                    broader.append(field.value()[2:])
                elif field.subfields[0] == 'a':
                    related.append(field.value())
        print('; '.join([' + '.join(maori),
                         ' + '.join(instead),
                         ' + '.join(related),
                         ' + '.join(broader)]))
```
